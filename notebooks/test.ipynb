```python
import fitz

def extract_pdf_urls(file_path):
    doc = fitz.open(file_path)
    urls = []
    for page in doc:
        links = page.get_links()
        if links:
            for link in links:
                uri = link.get("uri", None)
                if uri:
                    urls.append(uri)
    doc.close()

    return urls
```
